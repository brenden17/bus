```python
import pandas as pd

bus_df = pd.read_csv('busall.tsv', sep='\t')
bus_df = bus_df[bus_df['stationid']!='0']
bus_df = bus_df[bus_df['stationid']!='미정차']
# bus_df.columns

# bus_df.columns
busstop_df = bus_df.drop_duplicates(cols='stationid', take_last=True)
countstaion_s = bus_df['stationid'].value_counts()
# print(countstaion_s.values)
d = pd.DataFrame([countstaion_s.index, countstaion_s.values])
dd = d.T
dd.columns = ['stationid', 'count']
dd = pd.merge(bus_df, dd, on='stationid')

# group bus no
print(bus_df.head())
grouped = bus_df.groupby(['stationid'])
d = grouped.groups
print(d)
busno_df = pd.DataFrame([[k, ', '.join(map(str, d[k]))] for k, v in d.items()])
busno_df.columns = ['stationid', 'bus no']

busno_df = pd.merge(dd, busno_df, on='stationid')

busno_df
```
